```python
import pandas as pd
```

```python
staj_dwh = pd.read_csv('dwh.txt')
staj_dwh
```

```python
#Общая сумма стажа в DWH
s_dwh = staj_dwh.P_DAY1.sum()
print(s_dwh)
```

```python
#Сумма по кодам
staj_dwh_ = staj_dwh.groupby('CODE').P_DAY1.sum().reset_index()
staj_dwh_.rename(columns={'P_DAY1': 'staj_dwh'}, inplace=True)
staj_dwh_
```

3. Запускаем запрос по снилсу 70985583 [stg.sql](stg.sql)
4. Результат [stg.txt](stg.txt)

```python
# Подсчет стажа на STG011

staj_stg = pd.read_csv('stg.txt')
staj_stg
```

```python
staj_stg['PERIOD_BEGIN'] = pd.to_datetime(staj_stg['PERIOD_BEGIN'], format='%Y-%m-%dT%H')
staj_stg['PERIOD_END'] = pd.to_datetime(staj_stg['PERIOD_END'], format='%Y-%m-%dT%H')
staj_stg.loc[:, ['PERIOD_BEGIN', 'PERIOD_END', 'CODE', 'FLG_TAKE', 'FLG_STAJ_AGR']].head()
```

```python
def staj(row):
    for date in row:
        staj = row.PERIOD_END - row.PERIOD_BEGIN
    return staj
```

```python
staj_stg['staj_stg'] = staj_stg.apply(staj, axis=1)
staj_stg_ = staj_stg.loc[:, ['staj_stg', 'PERIOD_BEGIN', 'PERIOD_END', 'CODE', 'FLG_TAKE', 'FLG_STAJ_AGR']]
staj_stg_
```

```python
#Общая сумма стажа в STG
s_stg = staj_stg_.staj_stg.sum()
print(s_stg)
```

```python
#Сумма по кодам
staj_stg__ = staj_stg.groupby('CODE').staj_stg.sum().reset_index()
staj_stg__
```

```python
# Объединяем таблицу со стажем из DWH и STG
staj_ = staj_dwh_.merge(staj_stg__, how='left', on='CODE')
staj_
```

#### Фильтруем FLG_TAKE = 1 FLG_STAJ_AGR = 1

```python
staj_stg_filtered = staj_stg_.loc[(staj_stg_.FLG_TAKE == 1) & (staj_stg_.FLG_STAJ_AGR == 1)]
staj_stg_filtered
```

```python
#Общая сумма отфильтрованного стажа в STG
s_stg_f = staj_stg_filtered.staj_stg.sum()
print(s_stg_f)
```

```python
#Стаж отфильрованный по кодам
staj_stg_filtered_ = staj_stg_filtered.groupby('CODE').staj_stg.sum().reset_index()
staj_stg_filtered_.rename(columns={'staj_stg': 'staj_stg_filtered'}, inplace=True)
staj_stg_filtered_
```

```python
# Объединяем таблицу со стажем из DWH , из STG и отфильтрованным STG
staj__ = staj_.merge(staj_stg_filtered_, how='left', on='CODE')
staj__
```
